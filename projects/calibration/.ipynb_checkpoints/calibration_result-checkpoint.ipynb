```python
data1 = np.loadtxt('results/original', delimiter=",")
data2 = np.loadtxt('results/constant', delimiter=",")
time = np.arange(0,len(Idata[:,0]))*0.006547764182235917
plt.plot(time,data1[:,9],'r')
plt.plot(time,data2[:,9],'g')
plt.plot(IR_data[:,0],IR_data[:,1],'b')
plt.legend(['simulation-temp_dependent','simulation-temp_independent','experiment'])
plt.show()
```

```python
data1 = np.loadtxt('results/original', delimiter=",")
data2 = np.loadtxt('results/adiabatic', delimiter=",")
time1 = np.arange(0,len(data1[:,0]))*0.006547764182235917
time2 = np.arange(0,len(data2[:,0]))*0.006547764182235917
plt.figure()
plt.plot(time1,data1[:,9],'r')
plt.plot(time2,data2[:,9],'g')
plt.plot(IR_data[:,0],IR_data[:,1],'b')
plt.legend(['isothermal','adiabatic','experiment'])
plt.show()
```

```python
data = np.loadtxt('results/adiabatic_substrate', delimiter=",")
time = np.arange(0,len(data[:,0]))*0.006547764182235917
substrate_data = np.loadtxt('experiments/substrate.csv', delimiter=",")
substrate_data[:,1:5] = substrate_data[:,1:5]+273.15
plt.figure()
plt.plot(time,data[:,0])
plt.plot(time,data[:,1])
plt.plot(time,data[:,2])
plt.plot(substrate_data[:,0],substrate_data[:,1],'--')
plt.plot(substrate_data[:,0],substrate_data[:,2],'--')
plt.plot(substrate_data[:,0],substrate_data[:,3],'--')
plt.plot(substrate_data[:,0],substrate_data[:,4],'--')
plt.show()
```
